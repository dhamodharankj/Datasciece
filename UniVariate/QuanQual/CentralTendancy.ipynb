```python
univariate=pd.DataFrame(index=['Mean','Median'],columns=quan)
for columnName in quan:
    univariate[columnName]['Mean']=dataset[columnName].mean()
    univariate[columnName]['Median']=dataset[columnName].median()
```

```python
univariate
```

```python
univariate=pd.DataFrame(index=['Mean','Median','Mode'],columns=quan)
for columnName in quan:
    univariate[columnName]['Mean']=dataset[columnName].mean()
    univariate[columnName]['Median']=dataset[columnName].median()
    univariate[columnName]['Mode']=dataset[columnName].mode()
```

```python
univariate
```

```python
univariate=pd.DataFrame(index=['Mean','Median','Mode'],columns=quan)
for columnName in quan:
    univariate[columnName]['Mean']=dataset[columnName].mean()
    univariate[columnName]['Median']=dataset[columnName].median()
    univariate[columnName]['Mode']=dataset[columnName].mode()[0]
```
