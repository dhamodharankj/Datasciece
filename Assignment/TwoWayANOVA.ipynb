```python
import numpy as np
import pandas as pd
from scipy import stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
my_data=pd.read_csv('Preprocessed.csv')
my_data
```

```python
my_data.isna().sum()
```

```python
#perform One-Way ANOVA
model=ols('salary ~ degree_t',data=my_data).fit()
aov_table=sm.stats.anova_lm(model,type=2)
aov_table
```

```python
model=ols('salary ~ degree_t+specialisation',data=my_data).fit()
aov_table=sm.stats.anova_lm(model,type=2)
aov_table
```
