```python
import numpy as np
import pandas as pd
```

```python
raw_df=pd.read_csv('Zomato_Chennai_Listing_2020.csv')
raw_df
```

```python
def locsplit(x):
    if len(x.split(','))==2:
        return x.split(',')[1].replace(' ','')
    else:
        return x
```

```python
raw_df['Location_2'] = raw_df['Location'].apply(lambda x: locsplit(x))
```
